```python
from leetscrape import GetQuestionsList
import pandas as pd
import webbrowser
import json
```

```python
ls = GetQuestionsList()
ls.scrape()
qinfo = ls.questions
```

```python
qinfo= qinfo[qinfo['paidOnly']==False]
qinfo
```

```python
quesSubdir = list(qinfo['titleSlug'])
```

```python
def getUrl(quesPath):
    baseUrl='https://leetcode.com/problems/'
    return baseUrl+quesPath+'/solutions'
```

```python
urlArray = list(map(getUrl,quesSubdir))
```

```python
len(urlArray)
```

```python
qinfo = qinfo.assign(solutionUrl=urlArray)
```

```python
qinfo.head()
```

```python
qinfo['categorySlug'].fillna('nil',inplace=True)
```

```python
jsonRecord = json.dumps(qinfo.to_dict('records'))
```

```python
jsonFile = open("questions.json", "w")
jsonFile.write(jsonRecord)
jsonFile.close()
```
